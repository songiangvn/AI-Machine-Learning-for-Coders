## Getting Started with Machine Learning

 Consider these two sets of numbers:

 X = –1, 0, 1, 2, 3, 4

 Y = –3, –1, 1, 3, 5, 7

 There’s a relationship between the X and Y values (for example, if X is –1 then Y is –3,
 if X is 3 then Y is 5, and so on). Can you see it?

```python
import tensorflow as tf
import numpy as np
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense

"""
Định nghĩa 1 neural network đơn giản với 1 layer và 1 neuron
"""
linear = Dense(units=1, input_shape=[1])
model = Sequential([linear])
"""
Định nghĩa các layer bằng Sequential()
Inside Sequential, specify what each layer looks like
Với code này chỉ có 1 layer là Dense - layer phổ biến nhất, layer mà tất cả các neuron
của layer này đều connect tới tất cả các neuron ở next layer

units đại diện cho số lượng đầu ra - ở đây chỉ ra 1 y nên units = 1
Ở đây Dense Layer có units = 1 => chỉ có 1 dense layer với 1 neuron trong layer
Cuối cùng ở đây input chỉ là 1 đầu vào x nên input_shape chỉ là single value
"""
model.compile(optimizer='sgd', loss='mean_squared_error')
"""
Ở đây opimizer là cách mà mô hình optimize: sgd là stochastic gradient descent
Here's the part where the math comes in
loss function - truyền vào để model đánh giá how good or how bad that guess is
optimizer - computer can then make another guess, based on how the loss function went
"""

xs = np.array([-1.0, 0.0, 1.0, 2.0, 3.0, 4.0], dtype=float)
ys = np.array([-3.0, -1.0, 1.0, 3.0, 5.0, 7.0], dtype=float)

model.fit(xs, ys, epochs=500)
"""
epochs là try 500 times
Ở mỗi tries, computer sẽ guess the relationship và đánh giá độ good or bad
thông qua loss function
Sau đó optimizer sẽ dựa vào tính "good or bad" đó để chuyển tới next guess
Quá trình sẽ được lặp lại và result sẽ better and better
"""
```

```python
# Xuất kết quả dự đoán với x = 10
print("Kết quả dự đoán với x = 10 là:" ,model.predict(np.array([10.0])))
```

Ở đây mọi người có thể thấy 2 khái niệm khá mới với người mới học là hàm tối ưu và hàm mất mát ha. Chúng là 2 thứ hầu như sẽ xuất hiện xuyên suốt trong tất cả các mô hình, mình sẽ không đi sâu về toán nên là theo mình mọi người chỉ cần hiểu đơn giản là vầy:

- Hàm mất mát giúp mô hình tính số lỗi sai mà mình đã mắc phải, từ đó có thể nhận thức được kết quả từ quá trình học đã tốt hơn và giống với mục tiêu đề ra chưa. Qua đó làm tiền đề cho bước học tiếp theo.
- Hàm tối ưu giúp mô hình tối ưu hóa kết quả, giảm thiểu số lỗi về mức thấp nhất từ đó đạt được hiệu quả chính xác cao hơn. Nó giống như cách học của con người chúng ta vậy, con người có nhiều phương pháp, cách học khác nhau rồi luyện tập, sửa lỗi sai không ngừng cho đến khi đúng thì mô hình cũng thế.

## Seeing what the network learned(Xem xét các trọng số và độ lệch - weight và bias(độ lệch) mà mô hình đã học được)

As mentioned in the previous section, neurons have weight and bias
 parameters that they learn, which makes a single neuron fine for learning a relation
ship like this: namely, when Y = 2X – 1, the weight is 2 and the bias is –1. With
 TensorFlow we can actually take a look at the weights and biases that are learned,
 with a simple change to our code like this:

```python
print("Here is what I learned: {}".format(linear.get_weights()))
# in ra để check w với b xem line fit với data x và y đầu vào
```

In this case the relationship that model predict is: Y = 1.9976441X - 0.99269617

 This is pretty close to what we’d expect (Y = 2X – 1), and we could argue that it’s even
 closer to reality, because we are assuming that the relationship will hold for other
 values

## Summary:

Những kiến thức nắm bắt được:

- Dense(units, input_shape) để khởi tạo lớp tuyến tính ( linear).
